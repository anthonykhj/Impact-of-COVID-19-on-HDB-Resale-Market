# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Datasets

```python
df = pd.read_csv("resale_flat_prices_based_on_registration_date_from_jan_2017_onwards.csv")
df
```

```python
df1 = pd.read_csv("owid-covid-data.csv")
df1
```

# Filtering Datasets

```python
df_resale = df[df["month"].str.contains('2020')]
df_resale
```

```python
df1 = df1[["location", "date", "new_cases"]]
df_covid = df1[df1["location"].str.contains("Singapore") & df1["date"].str.contains("2020")]
```

# Checking for Null Values

```python
df_resale.isnull().sum() # 0 null values
```

```python
df_covid.isnull().sum() # 0 null values
```

# Setting Seaborn Color Palettes 

```python
# Create an array with the colors you want to use
colors = ["#BF2034", "#003478","#E6B04B", "#943173", "#CFE2F3", "#444444", "#a9a9a9"]
# Set your custom color palette
sns.set_palette(sns.color_palette(colors))
```

# Line Chart: COVID-19 Numbers

```python
# Converting the "date" column to datetime format
df_covid["date"] = pd.to_datetime(df_covid["date"])

# Extracting the month name and storing it in a new column called "month"
df_covid["month"] = df_covid["date"].dt.strftime("%b").str.upper()

df_covid
```

```python
# Plotting Daily New COVID-19 Cases in 2020
df_covid.plot(x="month", y="new_cases", kind="line", color="#003478")
plt.xlabel("Month")
plt.ylabel("Daily New Cases")
plt.title("Daily New COVID-19 Cases in 2020")
plt.show()
```

# Bar Chart: Transaction Volume of HDB Resale Flats (2019 VS 2020)

```python
# Extracting 2019 and 2020 values
df_resale_comparison = df[df["month"].str.contains('2019|2020')]

# Extracting the years and putting it into a new column called 'years'
df_resale_comparison['year'] = pd.DatetimeIndex(df_resale_comparison['month']).year

# Creating a new dataframe that compares the transaction volume of 2019 and 2020
df_transvol_comparison = df_resale_comparison.groupby('year').size().reset_index(name='transaction_volume')
df_transvol_comparison
```

```python
sns.barplot(x='year', y='transaction_volume', data=df_transvol_comparison)
plt.xlabel('Year')
plt.ylabel('Transaction Volume')
plt.title('Transaction Volume of HDB Resale Flats (2019 VS 2020)')
plt.show()
```

```python
# Extracting the months of 2020 and their respective transaction volumes
df_transvol = df_resale.groupby('month').size().reset_index(name='transaction_volume')

# Replacing the values
df_transvol['month'].replace(['2020-01', '2020-02', '2020-03', '2020-04', '2020-05', '2020-06', '2020-07', '2020-08', '2020-09', '2020-10', '2020-11', '2020-12'],
                            ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'], inplace=True)
df_transvol
```

```python
sns.barplot(x='month', y='transaction_volume', data=df_transvol, color="#003478")
plt.xlabel('Month')
plt.ylabel('Transaction Volume')
plt.title('Transaction Volume of HDB Resale Flats (2020)')
plt.show()
```

# Pie Chart: Transaction Volume for Each Flat Type

```python
# Extracting the flat types and their respective transaction volumes
df_flat_types = df_resale.groupby('flat_type').size().reset_index(name='transaction_volume')
df_flat_types
```

```python
# Creating a new dataframe that grouped '1 ROOM' and 'MULTI-GENERATION' together because the values were too small individually
oneroom_multigen = pd.DataFrame({'flat_type': ['1 ROOM + MULTI-GENERATION'], 'transaction_volume': [df_flat_types.loc[0, 'transaction_volume'] + df_flat_types.loc[6, 'transaction_volume']]})

# Dropping the original '1 ROOM' and 'MULTI-GENERATION' rows
df_flat_types = df_flat_types.drop([0,6])

# Adding the new dataframe to the original dataframe and sorting them alphabetically
df_flat_types = pd.concat([df_flat_types, oneroom_multigen], ignore_index=True)
df_flat_types = df_flat_types.sort_values('flat_type')
df_flat_types
```

```python
# Plotting the pie chart
explode = (0.5, 0, 0, 0, 0, 0)
plt.pie(df_flat_types['transaction_volume'], autopct='%1.1f%%', startangle=175, explode=explode, pctdistance = 0.85, textprops={'fontsize':10, 'color':'white'})

# Converting the pie chart into a donut chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.legend(df_flat_types['flat_type'], bbox_to_anchor=(1.2, 0.5))
```

#### Note: '1 ROOM + MULTI-GENERATION' is LESS THAN 0.1%, hence it is not exactly visible on the pie chart.

# Boxplot: HDB Resale Price for Each Flat Type

```python
# Boxplot for Price against each flat_type
resale_price_by_flat_type = sns.boxplot(y='flat_type',x='resale_price', data=df_resale, order=sorted(df_resale['flat_type'].unique()), palette = colors)
plt.ylabel('Flat Type')
plt.xlabel('Resale Price ($000,000)')
plt.title('Price for Each Flat Type')
plt.show()
```

# Horizontal Bar Chart:
# Max & Min Resale Price & Corresponding Town for Each Flat Type

```python
# Creating a new dataframe that stores the max resale price and min resale price under each flat type
df_max_min_price = pd.DataFrame(df_resale.groupby('flat_type').resale_price.max())
df_max_min_price = df_max_min_price.rename(columns={'resale_price':'max_resale_price'})
df_max_min_price['min_resale_price'] = df_resale.groupby('flat_type').resale_price.min()

# Adding the corresponding towns to the max and min resale price for each flat type
df_max_min_price['max_resale_town'] = df_resale.loc[df_resale.groupby('flat_type').resale_price.idxmax(), 'town'].values
df_max_min_price['min_resale_town'] = df_resale.loc[df_resale.groupby('flat_type').resale_price.idxmin(), 'town'].values
df_max_min_price
```

```python
# Creating the bar chart
df_max_min_price[['min_resale_price', 'max_resale_price']].plot(kind='barh')
plt.xlabel('Resale Price ($000,000)')
plt.ylabel('Flat Type')


# Adding the town labels to the bars
for i, (min_town, max_town) in enumerate(zip(df_max_min_price['min_resale_town'], df_max_min_price['max_resale_town'])):
    plt.text(min(df_max_min_price.iloc[i, 0:2]), i, min_town, ha='left', va='top', fontsize='x-small')
    plt.text(max(df_max_min_price.iloc[i, 0:2]), i, max_town, ha='left', va='bottom', fontsize='x-small')
    
plt.legend(['Minimum Resale Price', 'Maximum Resale Price'])
plt.margins(x=0.3)
plt.grid(False)
plt.show()
```
